**Para este ejercicio debe debe realizar lo siguiente:**

- Crear un diccionario de nombre dic_aminoacidos, donde se tenga la información de la tabla que se encuentra en http://www.russelllab.org/aas/aas.html. 
- Utilice como "key" los valores de la columna "AA" y las columnas "Hits" y "%Abundance" como "values". 
- Luego de crear el diccionario, extraiga los valores de "key" a una - lista de nombre "aminoacidos"
- Extraiga los valores de "value" a una lista de nombre "propiedades

````python
dic_aminoacidos = {'Trp':['1550','1.34'],'Cys':['1905','1.76'],'His':['1695','2.26'],'Arg':[ '3093', '5.20'],
              'Met':['1102','2.32'],'Tyr':['1525','3.25'],'Gln':['1830','3.96'],'Phe':['1616','4.12'],
              'Asp':['1691','5.12'],'Pro':['1598','5.00'],'Ala':['2018','7.34'],'Asn':['1103','4.57'],'Lys':['1366','5.81'],'Glu':['1471','6.22'],'Gly':['1386','6.89'],
               'Thr':['1147','5.85'],'Val':['1225','6.48'],'Ile':['1033','5.76'],'Ser':['1206','7.38'],'Leu':['1127','9.36']} 
dic_aminoacidos.keys()
aminoacidos = list()
for i in dic_aminoacidos.keys():
    aminoacidos.append(i)
print(aminoacidos)
propiedades = list() 
for e in dic_aminoacidos.values():
    propiedades.append(e)
print(propiedades)
````
